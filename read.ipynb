```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel('Analytic_test.xlsx', header=1, index_col=0)
df.info()
```

```python
info_colomns = {}
for col in df.columns:
    info_colomns[col] = df[col].nunique()
pd.Series(info_colomns)
```

```python
pd.set_option('display.max_columns', None)
df.head()
```

```python
def _search(x):
    try:
        x = int(float(x))
    except TypeError:
        return np.nan
    except ValueError:
        return np.nan
    return x

for col in list(df.columns)[1:]:
    df[col] = df[col].apply(_search)
df.info()
```

```python
df = df.dropna()
for col in list(df.columns)[1:]:
    df[col] = df[col].apply(_search)
df.info()
```
